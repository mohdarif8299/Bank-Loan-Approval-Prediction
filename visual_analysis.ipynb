```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bank_data.csv')
```

```python
#Displaying the loan_status count
loan_status_counts = df['loan_status'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(loan_status_counts.index, loan_status_counts.values, color=['blue', 'orange'])
plt.title('Histogram of Target Variable (Approved/Rejected)')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.show()
```

```python
# Displaying the boxplots for analyzing the numeric features
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

for var, ax in zip(['income_annual','loan_amount','residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value'], axes.flatten()):
    sns.boxplot(x='loan_status', y=var, data=df, ax=ax, palette='muted')
    ax.set_title(f'{var} by Loan Status')
    ax.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```

```python
# Using Count Plots for Analyzing Categorical Features
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
cat_vars = ['dependents', 'education', 'self_employed']

for var, ax in zip(cat_vars, axes.flatten()):
    sns.countplot(x=var, data=df, ax=ax,palette='viridis')
    ax.set_title(f'Count of {var}')

plt.tight_layout()
plt.show()
```
